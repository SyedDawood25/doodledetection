# Preparing a single dataset

```python
samples = np.concatenate((airplane, axe, basketball, bicycle, camera, diamond, donut, envelope, fish, helicopter), axis = 0)

y_airplane = np.zeros(len(airplane), dtype = np.uint)
y_axe = np.ones(len(axe), dtype = np.uint)

y_basketball = np.empty(len(basketball),dtype = np.uint)
y_basketball.fill(2)

y_bicycle = np.empty(len(bicycle),dtype = np.uint)
y_bicycle.fill(3)

y_camera = np.empty(len(camera),dtype = np.uint)
y_camera.fill(4)

y_diamond = np.empty(len(diamond),dtype = np.uint)
y_diamond.fill(5)

y_donut = np.empty(len(donut),dtype = np.uint)
y_donut.fill(6)

y_envelope = np.empty(len(envelope),dtype = np.uint)
y_envelope.fill(7)

y_fish = np.empty(len(fish),dtype = np.uint)
y_fish.fill(8)

y_helicopter = np.empty(len(helicopter),dtype = np.uint)
y_helicopter.fill(9)

labels = np.concatenate((y_airplane, y_axe, y_basketball, y_bicycle, y_camera, y_diamond, y_donut, y_envelope, y_fish, y_helicopter), axis = 0 )

print(f"Total: {len(samples)}")
```

# Visualizing the Target Distribution

```python
pd.DataFrame(labels).value_counts().plot(kind = 'bar' ,cmap='Spectral')
plt.xlabel('classes')
plt.ylabel('No. of samples per class')
```

# Shuffling the Samples Randomly

```python
np.random.seed(42)
shuffled_idx = np.random.permutation(len(samples))

samples = samples[shuffled_idx]
labels = labels[shuffled_idx]
```

```python
df = pd.DataFrame(samples)
df.head()
```

# Data Quality Assessment


## Visualizing the pixel distribution per image

```python
import seaborn as sb

fig, axes = plt.subplots(2,2, figsize = (8,8))

sb.histplot(samples[0] , ax= axes[0,0])
sb.histplot(samples[1] , ax= axes[0,1])
sb.histplot(samples[2] , ax= axes[1,0])
sb.histplot(samples[3] , ax= axes[1,1])
```

# Converting data to mini batches


```python
batches = 100
samples_per_batch = len(samples) // batches
batches_arr = []

for i in range(0,(batches*samples_per_batch) , samples_per_batch):
    batches_arr.append(samples[i:i+samples_per_batch].copy())

batches_arr[len(batches_arr)-1] = np.concatenate((samples[batches*samples_per_batch : ].copy(),batches_arr[len(batches_arr)-1]))

batches_arr_l = []

for i in range(0,(batches*samples_per_batch) , samples_per_batch):
    batches_arr_l.append(labels[i:i+samples_per_batch].copy())

batches_arr_l[len(batches_arr_l)-1] = np.concatenate((labels[batches*samples_per_batch : ].copy(),batches_arr_l[len(batches_arr_l)-1]))
```

```python
len(batches_arr_l)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()

for i in range(len(batches_arr)):
    batches_arr[i] = scaler.fit_transform(batches_arr[i])

pd.DataFrame(batches_arr[0]).head()
```

# Preparing Train, Validation And Test Sets

```python
train = batches_arr[0:70]
val = batches_arr[70:80]
test = batches_arr[80:]

train_l = batches_arr_l[0:70]
val_l = batches_arr_l[70:80]
test_l = batches_arr_l[80:]

val_batch = val[0] 
val_batch_l = val_l[0] 

test_batch = test[0]
test_batch_l = test_l[0]

for i in range(1,len(val)):
   val_batch = np.concatenate((val[i],val_batch))
   val_batch_l = np.concatenate((val_l[i],val_batch_l))

for i in range(1,len(test)):
   test_batch = np.concatenate((test[i],test_batch))
   test_batch_l = np.concatenate((test_l[i],test_batch_l))

print(f"Training Size : {samples_per_batch*70}")
print(f"Validation Size : {len(val_batch)}")
print(f"Test Size : {len(test_batch)}")

print(f"Total : {sum([samples_per_batch*70,len(val_batch),len(test_batch)])}")
```

# Saving the processed data into a compressed file

```python
np.savez_compressed('data.npz', train, train_l, val_batch, val_batch_l, test_batch, test_batch_l)
```
